```python
from receptive_fields.util.experiment_setup import load_dataset
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
```

```python
dataset = load_dataset(dataset="imagenette", train_data=True, enable_img_transforms=True, img_size=(320,240))
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)
```

```python
fig, axes = plt.subplots(4,8, figsize=(20,8))
for ax, img in zip(axes.flat, iter(dataloader)):
    ax.imshow(img[0][0].movedim(0,2))
    ax.axis('off')
plt.show()
```

```python
max_width=0
min_width=9999999
max_height=0
min_height=9999999
for img, label in dataloader:
    max_width = max(max_width,img.shape[-2])
    max_height = max(max_height,img.shape[-1])
    min_width = min(min_width,img.shape[-2])
    min_height = min(min_height,img.shape[-1])
```
